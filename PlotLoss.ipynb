```python
def plotloss(fname,factor):
    if os.path.isfile(fname):
        m=io.loadmat(fname)
        m['loss']=m['loss'][:,1:].reshape((-1,factor))
        m['t'] = np.arange(1,1+m['loss'].shape[0])*factor
        plt.plot(m['t'],np.mean(m['loss'],axis=1),'.-');
        return m
```

```python
factor=1000
fs = ['fibsem2/loss.mat',
      'fibsem3/loss.mat',
      'fibsem3_adam/loss.mat',
      'fibsem3_rmsprop/loss.mat',
      'fibsem4/loss.mat',
      'fibsem5/loss.mat',
     ]

plt.subplot(121)
ms = [plotloss(f,factor) for f in fs]
legstr = [os.path.split(f)[0] for f in fs]
plt.grid();
plt.legend(legstr);
plt.xlim((0,8000))

plt.subplot(122)
ms = [plotloss(f,factor) for f in fs]
legstr = [os.path.split(f)[0] for f in fs]
plt.grid();
plt.legend(legstr);
```
